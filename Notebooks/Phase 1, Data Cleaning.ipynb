Coloumn Name Cleaning

```python
print(f"Original Column Names:\n{data.columns.tolist()}")
```

```python
data.columns=(
    data.columns
    .str.strip()
    .str.replace(' ','_')
    .str.replace('[^A-Za-z0-9_]+', '', regex=True)
)

print(f"Cleaned Coloumn Names:\n{data.columns.tolist()}")
```

```python
import matplotlib.pyplot as plt

continuous_features=['AgeBaseline', 'CholesterolBaseline','CreatinineBaseline','eGFRBaseline','sBPBaseline','dBPBaseline','BMIBaseline']

for feature in continuous_features:
    plt.figure(figsize=(6,4))
    plt.hist(data[feature],bins=30,edgecolor='k')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import seaborn as sns 
corr_matrix=data.corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f",cmap="coolwarm",square=True)
plt.title("Correlation Matrix of Features")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of continuous features
continuous_features = [
    'AgeBaseline', 'CholesterolBaseline', 'CreatinineBaseline', 
    'eGFRBaseline', 'sBPBaseline', 'dBPBaseline', 'BMIBaseline'
]

for feature in continuous_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='EventCKD35', y=feature, data=data)
    plt.title(f'{feature} vs CKD Event (EventCKD35)')
    plt.xlabel('CKD Event (0=No, 1=Yes)')
    plt.ylabel(feature)
    plt.grid(True)
    plt.show()
```

```python
continuous_features = [
    'AgeBaseline', 'CholesterolBaseline', 'CreatinineBaseline', 
    'eGFRBaseline', 'sBPBaseline', 'dBPBaseline', 'BMIBaseline'
]

for feature in continuous_features:
    num_zeros = (data[feature] == 0).sum()
    print(f"{feature} - Zero count: {num_zeros}")

data[continuous_features].describe()
```
